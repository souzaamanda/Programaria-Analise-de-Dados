quartil: divede o dado em 4 partes(1/4 dos dados)

1º quartil(25%) é a metadade de baixo da mediana - O "limite inferior". 25% dos dados são menores ou iguais a este valor.

2º Quartil (50%) A Mediana. Divide o conjunto exatamente ao meio.

3º quartil (75%) é a metade de cima da mediana - O "limite superior". 75% dos dados são menores ou iguais a este valor.

Inter quatil =  valor que tem entreo 1º e o 3º

# Correlação

DESCREVE A RELAÇÃO ENTRE VARIÁVEIS
- Positiva: quando duas variaveis aumentam juntas
- Negativa: uma variavel aumenta enquanto a outra diminui

A correlação de Pearson (r) é uma medida estatística que quantifica a intensidade e a direção da relação linear entre duas variáveis quantitativas.
-  Essa medida varia de -1 a 1
- Valores proximo de 1 indica uma forte correlação Positiva
- Valores proximo de - 1 indica uma forte correlação Negativa
- Valor proximo de 0 indica que não há uma relação linear entre as variáveis

```python
correlacao_continua = dados['IDADE'].corr(dados['SALARIO'])
correlacao_continua
```

Correlação entre variáveis discretas

O V de Cramér (ou **coeficiente de Cramér**) é uma medida de associação estatística usada para avaliar a força da relação entre duas variáveis categóricas nominais, variando de 0 a +1. Baseado no teste Qui-quadrado, ele indica quão forte é a associação (0 = nenhuma, 1 = perfeita)


```python
# importanda função chi2 da biblioteca stats
# compara a distribuição observada na tabela cruzada com uma distribuição esperada se as duas variáveis fossem independentes uma da outra.
# Quanto maior adiferença entre a distribuição observada e esperada, maior será o valor do qui quadrado, o que indica uma associação mais forte entre as variáveis
from scipy.stats import chi2_contingency
```

```python
def cramer_coeficiente(coluna1, coluna2):
  tabela_cruzada = np.array(pd.crosstab(coluna1, coluna2))
  chi2 = chi2_contingency(tabela_cruzada)[0] # retorna apenas o primeiro valor [0]
  soma =  np.sum(tabela_cruzada) # retorna a soma de cada categoria da coluna 'nivel de ensino'
  mini = min(tabela_cruzada.shape)-1 # valor minimo da tabela -1 (parametro de formula)
  cramer = np.sqrt(chi2/(soma*mini))
  return cramer
```
